```python
numbers = [0, 1, 2, 3, 4, 5, 6]

print(numbers[2:5])    # 인덱스 2,3,4까지 - 출력: [2, 3, 4]
print(numbers[:3])     # 인덱스 0,1,2까지 - 출력: [0, 1, 2]
print(numbers[4:])     # 인덱스 4부터 끝까지 - 출력: [4, 5, 6]
print(numbers[::2])    # 인덱스 0부터 끝까지 2의 간격으로 - 출력: [0, 2, 4, 6] (간격이 2인 슬라이싱)
print(numbers[::-1])   # 끝부터 처음까지 1의 간격으로(=역순) - 출력: [6, 5, 4, 3, 2, 1, 0] (역순)
```

```python
# 리스트 컴프리헨션 사용
squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print(squares)  # 출력: [4, 16, 36, 64, 100]

# 리스트 컴프리헨션 미사용
squares = []
for x in range(1, 11):
    if x % 2 == 0:
        squares.append(x**2)
print(squares)  # 출력: [4, 16, 36, 64, 100]
```

```python
# 실습 문제 2 : 시험 점수 딕셔너리 분석
import numpy as np
scores = {"Alice": 88, "Bob": 95, "Charlie": 70, "Dave": 95, "Eve": 62}
t= list(scores.values())
answer1 = np.mean(t)
print(f"평균 점수: {answer1}")
maxim =max(t)
max_score = [f"{x}({y})" for x,y in scores.items() if y==maxim]
max_score_joined = ", ".join(max_score)
print(f"최고 득점자: {max_score_joined}")

dictionary_diff = {x:'Pass' if y>=70 else 'Fail' for x,y in scores.items()}
print(dictionary_diff)
```
