```python
# Q3. 논리 연산자의 반환값 (Return Value)
print([] or "Hello" and 0 or 7)
```

SQL에서와 마찬가지로 and가 or보다 높은 우선순위를 가진다.
결국 [] or("Hello" and 0) or 7 이런 식으로 해석이 된다.
이를 논리타입으로 해석해서 본다면,
False or ("Hello" and False) or 7이 된다.
True값 and False는 False이므로 결론적으로
False or False or 7이므로 7이 프린트 된다.

```python
# Q4. 중첩 조건 표현식 (Nested Conditional Expression)
score = 75
result = \
    "A" if score > 90 \
        else "B" \
        if score > 70 \
        else "C"
print(result)
```

위에 90으로 구분한 else에서 벌써 B로 판별났기에 밑에 IF는 의미가 없다.

```python
# Q6. 리스트와 조건문
items = [None]
if items:
    print("Empty" if not items[0] else "Full")
else:
    print("No List")
```

items는 None이 들어있는 리스트이다. 그렇기에 true이므로 else문이 아닌, if문으로 간다.

```python
# Q11. 복합 사용자 권한 및 보안 검사 시스템
user_data = {"id": "admin", "active": True, "strikes": 0}
login_attempt = {"ip_allowed": False, "is_vpn": True, "mfa_verified": True}

# 복잡한 조건 분기 시작
status = "Denied"

if user_data["active"] and not (user_data["strikes"] >= 3):
    if login_attempt["ip_allowed"] or (login_attempt["is_vpn"] and login_attempt["mfa_verified"]):
        if user_data["id"] == "admin":
            status = "Super_Access" if login_attempt["ip_allowed"] else "Limited_Admin_Access"
        else:
            status = "User_Access"
    else:
        status = "Security_Block"
else:
    status = "Account_Disabled"

final_status = "Access_Granted" if status in ["Super_Access", "User_Access"] else "Investigation_Required" if status == "Limited_Admin_Access" else "Access_Rejected"

print(final_status)
```

Status에서 "Limited_Admin_Access"가 나오고 이는
final_status에서 "Investigation_Required"로 된다.

```python
# Q12. 단락 평가(Short-circuit)와 가변 객체의 함정
data = []
flag = True
result = ""

# 리스트가 비어있는지 확인하고, 비어있다면 요소를 추가하는 것처럼 보이지만...
# 파이썬의 and/or 우선순위와 단락 평가를 주의깊게 보세요.
check = (flag or data.append("A")) and (len(data) > 0 or data.append("B"))

if data:
    if "B" in data:
        result = "Path 1"
    else:
        result = "Path 2"
else:
    if not check:
        result = "Path 3"
    else:
        result = "Path 4"

print(f"Data: {data}, Result: {result}")
print(check)
```

check은 data.append("B")를 시행하고나서 and에의해 아무 것도 없는 변수가 되어버린다.
대신 B가 append되었기에 Path 1이 선택된다.

```python
# Q13. 중첩 조건 표현식과 Truthy/Falsy의 연쇄
val_1 = 0
val_2 = "False"
val_3 = []
val_4 = "Python"

step_1 = val_1 or val_2 if not val_3 else val_3 or val_4
step_2 = val_4 and val_1 if val_1 else val_2 and val_4

if step_1 == "False" and not step_2:
    final_value = (val_1 or val_3 or "Case A")
elif step_1 and step_2:
    final_value = (val_4 and "Case B")
else:
    final_value = "Case C"

print(final_value)
```

step_1은 (1 or 2) if not 3 else (3 or 4)
여기서 3은 false이기에 1 or 2를 본다. 여기서 "False"의 값을 가져간다.
step_2는 (4 or 1) if 1 else (2 or 4)
여기서 1은 false이기에 2 or 4를 본다. 여기서 첫 true값인 "False"를 가져간다.
step_1, step_2는 true이기에 "Case B"
