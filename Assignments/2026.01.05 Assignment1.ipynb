#### ✅ `end` 예제 (4개)

```python
# 예제 1: 줄바꿈 대신 느낌표를 붙이고 줄바꿈은 직접 넣기
print("안녕", end="!!!\n")  # \n은 줄바꿈 문자입니다.
# 예제 2: 줄바꿈을 없애서 같은 줄에 이어서 출력
print("첫째", end=" ")
print("둘째", end=" ")
print("셋째")
# 예제 3: 점점 찍는 로딩 느낌
for i in range(5):
    print(".", end="")
print(" 완료!")
# 예제 4: end를 줄바꿈 대신 ' / '로 바꾸기
print("사과", end=" / ")
print("바나나", end=" / ")
print("포도")
```

#### ✅ f-string 예제 (5개)

```python
name = "적토마"
score = 95
print(f"{name}의 점수는 {score}점입니다.")
money = 1234567
print(f"가격은 {money:,}원입니다.")
pi = 3.141592
print(f"원주율은 {pi:.2f}입니다.")
# 계산 결과를 바로 넣을 수도 있습니다.
a, b = 7, 3
print(f"{a} + {b} = {a + b}")
# 소수 둘째 자리 + 콤마를 같이 쓸 수도 있습니다.
big = 1234567.891
print(f"큰 수: {big:,.2f}")
```

#### ✅ `try/except` 예제 (5개)

```python
# 예제 1: 정수 변환 실패를 잡기
s = input("정수를 입력하세요: ")
try:
    n = int(s)
    print("n * 2 =", n * 2)
except ValueError:
    print("숫자가 아니라서 변환에 실패했습니다!")
# 예제 2: 실수 변환 실패를 잡기
s = input("실수를 입력하세요(예: 3.14): ")
try:
    x = float(s)
    print("x + 1 =", x + 1)
except ValueError:
    print("실수로 바꿀 수 없는 입력입니다!")
# 예제 3: 0으로 나누기 에러도 잡기
s = input("정수 2개를 공백으로 입력하세요(예: 10 0): ")
try:
    a, b = map(int, s.split())
    print("a / b =", a / b)
except ValueError:
    print("입력 형식이 잘못되었습니다! 예: 10 3")
except ZeroDivisionError:
    print("0으로는 나눌 수 없습니다!")
# 예제 4: 콤마가 포함된 숫자 처리(간단 버전)
s = input("콤마 포함 숫자를 입력하세요(예: 1,234): ")
s2 = s.replace(",", "")  # 콤마 제거
try:
    n = int(s2)
    print("변환 성공:", n)
except ValueError:
    print("숫자로 바꿀 수 없습니다!")
# 예제 5: 빈 입력 처리(간단)
s = input("아무것도 안 치고 엔터도 눌러보세요: ")
if s == "":
    print("빈 입력입니다. 다시 입력이 필요합니다.")
else:
    print("입력한 값:", s)
```

#### ✅ `split()` 예제 (6개)

```python
print("10 20 30".split())
print("a,b,c".split(","))
print("Read#Analyze#Implement".split("#"))
# 연속 공백은 split()가 알아서 정리해 줍니다.
print("10   20    30".split())
# 하지만 split(',')은 공백을 자동으로 없애주지 않습니다.
print("10, 20, 30".split(","))
# 공백이 거슬리면 strip()으로 정리할 수 있습니다.
parts = "10, 20, 30".split(",")
clean = [p.strip() for p in parts]
print(clean)
```
