#### ✅ 자주 쓰는 패턴 예제 (8개)

```python
s = "abcdefg" #[start:end:step]
print("[::-1] 뒤집기:", s[::-1])
s = "abcdefg" #[start:end:step]
print("[1::2] 1번부터 2칸씩:", s[1::2])
s = "abcdefg"#[start:end:step]
print("[-5:] 뒤에서 5글자:", s[-5:])
s = "0123456789"#[start:end:step]
print("[::2] 짝수 인덱스:", s[::2])
s = "0123456789"#[start:end:step]
print("[1::2] 홀수 인덱스:", s[1::2])
s = "Python"#[start:end:step]
print("[::-2] 뒤에서 2칸씩:", s[::-2])
s = "Hello, World!" #[start:end:step]
print("[7:12] 부분 추출:", s[7:12])
s = "Korea"#[start:end:step]
print("[0:100] end가 넘어가도 괜찮음:", s[0:100])
```

#### 공백제거✅ 예제 (6개)

```python
text = "   hello   "
print("원본:", repr(text))
print("strip:", repr(text.strip()))
print("lstrip:", repr(text.lstrip()))
print("rstrip:", repr(text.rstrip()))
# 탭(	)도 공백처럼 제거됩니다.
text = "\t\tHi\t"
print("원본:", repr(text))
print("strip:", repr(text.strip()))
# 특정 문자도 제거할 수 있습니다(양쪽에서만)
text = "###title###"
print(text.strip("#"))
# (헷갈림) strip은 "가운데"는 못 지웁니다.
text = "a  b  c"
print(text.strip())   # 가운데 공백은 그대로
# (자주 쓰는 패턴) split 전에 strip
line = "   10, 20, 30   "
parts = [p.strip() for p in line.strip().split(",")]
print(parts)
# 입력을 받을 때 빈 문자열인지 확인하는 데도 씁니다.
s = "   "
print(s.strip() == "")
```

#### 대소문자 통일✅ 예제 (5개)

```python
s = "PyThOn"
print(s.lower())
print(s.upper())
# 검색을 할 때 대소문자 무시하고 싶다면?
text = "Hello Python"
keyword = "python"
print(keyword in text)            # False 가능
print(keyword in text.lower())    # True
# (헷갈림) lower/upper도 새 문자열입니다(원본은 안 바뀜)
s = "ABC"
s.lower()
print(s)
s = "ABC"
s = s.lower()
print(s)
# 사용자 입력 통일(예: YES/yes/Yes 모두 처리)
ans = input("yes/no 입력: ").strip().lower()
print("입력 정리:", ans)
print(ans == "yes")
```

#### 바꾸기✅ 예제 (6개)

```python
s = "I like apple"
print(s.replace("apple", "banana"))
# 여러 번 등장하면 모두 바뀝니다.
s = "ha ha ha"
print(s.replace("ha", "ho"))
# 특정 문자 제거(공백 제거 등)에도 쓸 수 있습니다.
s = "1,234,567"
print(s.replace(",", ""))
# (헷갈림) 원본은 안 바뀝니다.
s = "apple"
s.replace("a", "A")
print(s)
# 다시 저장해야 합니다.
s = "apple"
s = s.replace("a", "A")
print(s)
# 첫 번째만 바꾸고 싶으면 count 인자 사용(선택)
s = "banana"
print(s.replace("a", "A", 1))
```

#### 나누기/합치기✅ 예제 (8개)

```python
s = "a b c"
print(s.split())
s = "10,20,30"
print(s.split(","))
# 쉼표 뒤 공백이 있으면?
s = "10, 20, 30"
parts = s.split(",")
print(parts)
clean = [p.strip() for p in parts]
print(clean)
# join: 리스트를 문자열로 합치기
words = ["I", "like", "python"]
print(" ".join(words))
# join: 구분자를 바꿀 수도 있습니다.
words = ["2026", "01", "04"]
print("/".join(words))
# (헷갈림) split은 리스트, join은 문자열
s = "a,b,c"
parts = s.split(",")
print(parts, type(parts))
print("-".join(parts), type("-".join(parts)))
nums = [1, 2, 3]
print(",".join(map(str, nums)))
```

#### 개수/위치 찾기✅ 예제 (7개)

```python
s = "banana"
print(s.count("a"))
print(s.count("na"))
s = "Hello Python"
print(s.find("Python"))
print(s.find("Java"))  # 없으면 -1
# find 결과로 존재 여부 판단
s = "abc"
idx = s.find("d")
print(idx)
print(idx == -1)
# 특정 위치 이후에서 찾기(선택)
s = "ababa"
print(s.find("ba", 1))
# (헷갈림) find는 첫 번째 위치만
s = "ababa"
print(s.find("ba"))
# 두 번째 위치를 찾고 싶다면: 첫 번째 위치 다음부터 다시 find
s = "ababa"
first = s.find("ba")
second = s.find("ba", first + 1)
print(first, second)
# count와 find를 같이 사용하면 전처리에 도움이 됩니다.
s = "a,b,c,d"
print("쉼표 개수:", s.count(","))
print("첫 쉼표 위치:", s.find(","))
```
