```python
## 실습 문제 1 : 약수 구하기
num = int(input("정수를 입력하세요: "))
a= []
for i in range(1, num):
    if num%i == 0:
        a.append(i)
    else:
        continue
print(f"정수를 입력하세요: {num}")
print(f"{num}의 약수:",*a)
```

```python
## 실습 문제 2 : 문자열 뒤집기
in_sting=input()
print(f"문자열을 입력하세요: {in_sting}")
print(in_sting[::-1])
```

```python
## 실습 문제 3 : 숫자 맞히기 게임
answer= 4
b=1
while b>0:
    a=int(input())
    print(f"정수를 입력하세요 (1~10): {a}")
    if a==answer:
        print("정답입니다!")
        break
    else:
        print("틀렸습니다. 다시 입력하세요.")
```
