```python
# Q1. while 문과 단락 평가의 만남
i = 0
count = 0
while i < 10 and (i := i + 1):
    if i % 3 == 0:
        continue
    count += 1
    if count >= 5:
        break
print(i)
```

첫번째 loop i = 1 count = 1
두번째 loop i = 2 count = 2
세번째 loop i = 3 count = 2
네번째 loop i = 4 count = 3
다섯번째 loop i = 5 count = 4
여섯번째 loop i = 6 count = 4
일곱번째 loop i = 7 count = 5 break
print(i)는 7이 된다.

```python
# Q2. 중첩 반복문과 break의 범위
stars = 0
for i in range(3):
    for j in range(10):
        if j > i:
            break
        stars += 1
print(stars)
```

break

```python
# Q3. 리스트 컴프리헨션 내의 조건문
nums = [1, 2, 3, 4, 5]
result = [x * 2 if x % 2 == 0 else x * 3 for x in nums if x > 2]
print(result)
```

각각 nums의 값을 2이하는 제외하고 계산한 값이 나오게 된다.  for x in X의 뒤에는 filter condition인 것이다.

```python
# Q4. 딕셔너리 순회와 값 수정
data = {"A": 1, "B": 2, "C": 3}
res = 0
for key in data:
    if key == "B":
        continue
    res += data[key]
else:
    res *= 2
print(res)
```

else는 for문이 끝난 이후에 진행이 되는 구조이다.

```python
# Q6. 2차원 리스트의 리스트 컴프리헨션
matrix = [[i for i in range(2)] for j in range(3)]
print(matrix)
```

```python
# Q7. while-else 구조의 이해
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
else:
    print("Done")
print(n)
```

while문에서 else는 완벽히 끝이 나야지 실행이되는 조건문이다. 그렇지 않으면 출력이안된다.

```python
# Q8. 문자열 순회와 continue
word = "PYTHON"
result = ""
for char in word:
    if char in "AEIOU":
        continue
    result += char
print(result)
```

```python
# Q9. 무한 루프와 조건 제어
x = 1
while True:
    x <<= 1  # x에 2를 곱하는 것과 같음
    if x > 50:
        break
    if x % 3 == 0:
        x += 1
print(x)
```

x는 2배가 계속되므로 64에서 멈추게 된다.
x<<=1의 의미는 binary코드를 한글자씩 미룬다는 뜻이다.
이는 2배를 하는 것과 같다.

```python
# Q10. 리스트 컴프리헨션의 중첩
colors = ["Red", "Blue"]
items = ["Cup", "Pen", "Bag"]
combined = [f"{c}-{i}" for c in colors for i in items if "e" in c or "a" in i]
print(len(combined))
```

조합은 6개가 나오며 전부 or조건을 만족한다.
