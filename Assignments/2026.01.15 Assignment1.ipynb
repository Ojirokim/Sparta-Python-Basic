### 문제 1

학생의 이름과 나이를 저장하는 `Student` 클래스를 만들어 보세요.

**조건**

- `__init__` 메서드를 사용하세요.
- 이름(name)과 나이(age)를 속성으로 저장하세요.
- 학생의 정보를 출력하는 `print_info()` 메서드를 만드세요.

**출력 예시**

```
이름: 민수, 나이: 15
```

````python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def print_info(self):
        print(f"이름: {self.name}, 나이: {self.age}")
Hi = Student("민수", 15)
Hi.print_info()
````

### 문제 2

사각형을 표현하는 `Rectangle` 클래스를 만들어 보세요.

**조건**

- 가로(width), 세로(height)를 인스턴스 변수로 가집니다.
- 이 인스턴스 변수들은 밖에서 직접 접근이 불가능합니다. (힌트 : `__` )
- 사각형의 넓이를 반환하는 `get_area()` 메서드를 만드세요.
- 메서드를 호출하여 넓이를 출력하세요.

**힌트**

- 넓이 = 가로 × 세로

````python
class Rectangle:
    def __init__(self, width, height):
        self.width= width
        self.height= height
    def get_area(self):
        return self.width*self.height
x= Rectangle(3,4)
x.get_area()
````

### 문제 4

아래 코드를 실행했을 때 출력 결과를 예측하고,

그 이유를 **클래스와 객체 관점에서 설명하세요.**

```python
class Counter:
    count = 0
    def add(self):
        self.count += 1

a = Counter()
b = Counter()

a.add()
a.add()
b.add()

print(a.count)
print(b.count)
```

**힌트**

- `count`는 클래스 변수인지, 인스턴스 변수인지 생각해 보세요.

a는 2 b는 1이 프린트 된다.
계속 class의 메소드가 돌아갈때마다 count가 증가하기 때문이다.

````python
class Counter:
    count = 0
    def add(self):
        self.count += 1

a = Counter()
b = Counter()

a.add()
a.add()
b.add()

print(a.count)
print(b.count)
````

### 문제 5

아래 코드의 출력 결과를 예측하고, 왜 그렇게 되는지 설명하세요.

```python
class Box:
    def __init__(self, items):
        self.items = items

box1 = Box([1, 2, 3])
box2 = box1

box2.items.append(4)

print(box1.items)
```

**힌트**

- `box1`과 `box2`가 가리키는 객체를 생각해 보세요.
- 리스트는 변경 가능한 객체(mutable)입니다.

box1의 아이템의 리스트도 바뀌어서 나오게 된다.
box1이 가리키는 주소가 box2와 같은데, 대상이 새로 생기지 않고 그 자리에서 바뀌기 때문이다.

````python
class Box:
    def __init__(self, items):
        self.items = items

box1 = Box([1, 2, 3])
box2 = box1

box2.items.append(4)

print(box1.items)
````

### 문제 6

아래 코드를 실행했을 때 **출력 결과를 정확히 예측**하고,

왜 그런 결과가 나오는지 **클래스 변수 / 인스턴스 변수 관점에서 설명**하세요.

```python
class Score:
    total = 0

    def __init__(self, value):
        self.value = value
        Score.total += value

    def add(self, x):
        self.value += x
        Score.total += x

a = Score(10)
b = Score(20)

a.add(5)
b.add(3)

print(a.value)
print(b.value)
print(Score.total)
```

**힌트**

- `self.value`는 어디에 저장되는 값인가요?
- `Score.total`은 언제, 어디에서 증가하나요?

a는 15 b는 23이 나오고 total은 합친 값 38이 나온다.
Score.total은 인스턴스를 수정하는 것이 아닌 class의 variable을 변경한다.

````python
class Score:
    total = 0

    def __init__(self, value):
        self.value = value
        Score.total += value

    def add(self, x):
        self.value += x
        Score.total += x

a = Score(10)
b = Score(20)

a.add(5)
b.add(3)

print(a.value)
print(b.value)
print(Score.total)
````

### 문제 7

아래 코드에는 **의도와 다르게 동작하는 부분**이 있습니다.

**의도**

- 각 학생 객체가 **자기만의 점수 리스트**를 가져야 한다.

코드를 실행했을 때 실제 출력 결과를 예측하고,

왜 이런 문제가 발생하는지 설명한 뒤, **올바르게 수정하세요.**

```python
class Student:
    scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)

```

**힌트**

- `scores`는 클래스 변수인가요, 인스턴스 변수인가요?
- 리스트는 어떤 특징을 가진 객체인가요?

두번의 출력 전부 90, 80으로 나올 것이다.
scores는 각 인스턴스에 대한 값이 아니라, 클래스의 변수이기에
변수에 들어가는 값이 계속 쌓이게 된다.

````python
class Student:
    scores = []

    def add_score(self, score):
        self.scores.append(score)

s1 = Student()
s2 = Student()

s1.add_score(90)
s2.add_score(80)

print(s1.scores)
print(s2.scores)
````
