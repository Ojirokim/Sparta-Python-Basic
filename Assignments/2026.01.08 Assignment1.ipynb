```python
from curses.ascii import isalpha

### 필수 1번: `if`는 거짓이면 else로 가는가?
x = 0
if x:
    print("A")
print("B")
```

0은 false이기 때문에, if문은 실행되지 않는다. 밑에 있는 Print()는 if문과 관계없기에 실행된다.

```python
### 필수예제 2번: `elif`는 모두 검사될까?
score = 95
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
```

if는 순서대로 진행되기에 위에서 조건에 만족하면 끝이난다.

```python
### 필수예제 3번 : 조건 순서가 바뀌면 결과가 바뀐다
age = 10
if age >= 0:
    print("사람입니다.")
elif age >= 18:
    print("성인입니다.")
```

의도대로 진행된다. 10이 들어가는 곳은 위의 조건밖에 없기 때문이다.

```python
### 필수예제 4번:  `and` / `or` 혼동
x = 5
if x > 0 or x < 10:
    print("OK")
```

위의 조건은 모든 x에 대해서 만족한다. or조건 때문에 0보다큰것, 10보다 작은 것 모두 조건에 포함되기 때문이다.

```python
### 필수예제 5번 : `not`의 대상은 어디까지인가?
x = 5
if not x > 3:
    print("A")
else:
    print("B")
```

Not은 같은 줄의 if조건에만 해당하고 다른 elif조건은 개별적으로 not을 작성해야한다.
이후 else에서는 차집합이기에 not의 결과가 중요해지긴한다.

```python
### 필수예제 6번 : 삼항 연산자
x = 0
result = "양수" if x else "0 또는 음수"
print(result)
```

x가 true value라면 앞의 값을 false value라면 뒤의 값을 반환

```python
### 필수예제 7번
x=int(input())
if x%3 ==0 and x%7!=0:
    print("3의 배수입니다")
elif x%3 !=0 and x%7 ==0:
    print("7의 배수입니다")
elif x%3 ==0 and x%7 ==0:
    print("3과 7의 공배수입니다.")
else:
    print("3과 7의 배수가 아닙니다.")
```
