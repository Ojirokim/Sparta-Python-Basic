### 문제 5

다음 리스트에서 짝수만 골라 새로운 리스트를 만들어 출력하세요.

`nums = [3, 10, 7, 8, 2, 9, 12]`

````python
nums = [3, 10, 7, 8, 2, 9, 12]
wawa = [x for x in nums if x%2==0]
print(wawa)
````

### 문제 6

다음 코드의 출력 결과를 예측하세요.

```python
a = [1, 2, 3]
b = a
b.append(4)
print(a)
print(b)
```

a와 b는 같은 리스트를 가리키고 있으며, append(4)는 그자리에서 변경을하기 때문에 a,b 모두 더해진 결과를 볼 수 있다.

````python
a = [1, 2, 3]
b = a
b.append(4)
print(a)
print(b)
````

### 문제 7

다음 코드의 출력 결과를 예측하세요.

```python
a = [1, 2, 3]
b = a[:]
b.append(4)
print(a)
print(b)

```

a는 처음 리스트를 가리키고, b는 새로운 리스트를 가리키고 있으니 b는 더해진 결과가 나오고 a는 그대로 나올 것이다.

````python
a = [1, 2, 3]
b = a[:]
b.append(4)
print(a)
print(b)
````

### 문제 9
