### 문제 1

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요.

```python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
```

위에 +=는 __iadd__로 change in place이다. 그렇기에 a와 b의 값이 변경된 것을 가리킨다.
밑에 = a+는 __add__로 변경한후 새로운 리스트를 할당하는 것이다.
그렇기에 a는 변경이 되지만, b는 이전의 리스트를 계속 가리키고 있기에 b는 그대로다.

````python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
````

### 문제 2

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..
```python
x = [1, 2, 3]
y = x[:]
x[0] = 99
print(x, y)

p = [1, 2, [3, 4]]
q = p[:]
p[2][0] = 999
print(p, q)
```

Shallow copy와 Deep copy의 개념에 대해서 보여주고 있는 문제이다.
shallow copy에서 list는 데이터에 대한 reference만 가지고 있다. 결국 p와 q는 1 reference, 2 reference, list reference를 가지고 있는 것이다. 또한 별도로 안에 리스트의 값인 3,4는 같이 공유하고 있는 것이다.
그렇기 때문에 shallow copy의 경우에 안에 리스트를 변경하게 되면 p와 q에 공유가 된다.
만약 deep copy를 사용한다면, p만 변경될 것이다.
```python
import copy
q = copy.deepcopy(p)
```

````python
x = [1, 2, 3]
y = x[:]
x[0] = 99
print(x, y)

p = [1, 2, [3, 4]]
q = p[:]
p[2][0] = 999
print(p, q)
````

### 문제 3

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..

```python
t = (1, 2, [3, 4])
t[2].append(5)
print(t)

# 아래 줄이 실행된다면 어떤 일이 일어나는지(출력/에러 포함) 정확히 쓰세요.
# t[2] += [6]
```

(1,2,[3,4,5]가 프린트 될 것이다.
이는 리스트와 튜플의 mutable immutable개념을 통해 확인할 수 있다. 튜플은 immutable이지만, 안에 있는 리스트는 mutable이기 때문에 변경이 가능한 것이다.
하지만, +=는 __iadd__ change in place이기 때문에 immutable인 튜플은 변경이 불가능하여 오류가 발생할 것이다.

````python
t = (1, 2, [3, 4])
t[2].append(5)
print(t)
t[2] += [6]
````

### 문제 4

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..

```python
lst = [0, 1, 2, 3, 4]
lst[1:4] = (9, 8)
print(lst)

lst = [0, 1, 2, 3, 4]
lst[1:4] = [9, 8]
print(lst)
```

슬라이싱의 대변에 들어오는 object가 tuple인지 list인지는 관계가 없다.
중요한 사실은 iterable이 들어와야한다는 사실이다.
python은 이 iterable에서 하나의 값씩 읽어와서 채우게 되는 것이다.
하지만, 주의해야하는 점은 이것이 slicing이기 때문에 가능하다는 점이다.
만약 index만 뽑는 경우 대변에 tuple, list가 들어오면 tuple, list 상태로 들어간다.
만약, slicing의 대변이 iterable이 아니라면, TypeError를 반환하게 된다.

````python
lst = [0, 1, 2, 3, 4]
lst[1:4] = (9, 8)
print(lst)

lst = [0, 1, 2, 3, 4]
lst[1:4] = [9, 8]
print(lst)

lst = [0, 1, 2, 3, 4]
lst[4] = [9, 8]
print(lst)

lst = [0, 1, 2, 3, 4]
lst[1:4] = 9
````

### 문제 7

````python
d = {}
d[[1, 2]] = "a"
print(d)
````

### 문제 9

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..

```python
d = {"a": 1, "b": 2, "c": 3}
k = d.keys()
d["d"] = 4
print(list(k))

# 아래 줄이 실행된다면 어떤 일이 일어나는지(출력/에러 포함) 정확히 쓰세요.
# for key in d:
#     d[key + "_x"] = d[key]
```

dict_keys class는 특별하게 사전에 대한 live view를 제공한다. 그렇기 때문에 d에서 변경점은 k에서도 live로 볼 수 있는 것이다.
만약 live view 대신 snapshot을 원한다면,
```python
k=list(d.keys())
```
위와 같이하면 k에 영향을 주지 않는 snap shot이 가능하다.

for문을 시행하면, key의 값에 _x를 더해서 새로운 key를 만들어낸다. 이는 이전의 값을 변화시키는 것은 아니기에 가능하다. 하지만, python이 iterable을 읽고 있는 와중에 사전의 사이즈를 바꾸게 된다면, 사고를 방지하기 위해서 오류를 반환한다.

````python
d = {"a": 1, "b": 2, "c": 3}
k = d.keys()
d["d"] = 4
print(list(k))

for key in d:
    d[key + "_x"] = d[key]
````
