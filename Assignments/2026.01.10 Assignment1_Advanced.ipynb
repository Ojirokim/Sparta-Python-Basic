### 문제 1

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요.

```python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
```

위에 +=는 __iadd__로 change in place이다. 그렇기에 a와 b의 값이 변경된 것을 가리킨다.
밑에 = a+는 __add__로 변경한후 새로운 리스트를 할당하는 것이다.
그렇기에 a는 변경이 되지만, b는 이전의 리스트를 계속 가리키고 있기에 b는 그대로다.

````python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
````

### 문제 2
