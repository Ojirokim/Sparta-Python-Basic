### 문제 1

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요.

```python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
```

위에 +=는 __iadd__로 change in place이다. 그렇기에 a와 b의 값이 변경된 것을 가리킨다.
밑에 = a+는 __add__로 변경한후 새로운 리스트를 할당하는 것이다.
그렇기에 a는 변경이 되지만, b는 이전의 리스트를 계속 가리키고 있기에 b는 그대로다.

````python
a = [1, 2, 3]
b = a
a += [4]
print(a, b)

a = [1, 2, 3]
b = a
a = a + [4]
print(a, b)
````

### 문제 2

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..
```python
x = [1, 2, 3]
y = x[:]
x[0] = 99
print(x, y)

p = [1, 2, [3, 4]]
q = p[:]
p[2][0] = 999
print(p, q)
```

Shallow copy와 Deep copy의 개념에 대해서 보여주고 있는 문제이다.
shallow copy에서 list는 데이터에 대한 reference만 가지고 있다. 결국 p와 q는 1 reference, 2 reference, list reference를 가지고 있는 것이다. 또한 별도로 안에 리스트의 값인 3,4는 같이 공유하고 있는 것이다.
그렇기 때문에 shallow copy의 경우에 안에 리스트를 변경하게 되면 p와 q에 공유가 된다.
만약 deep copy를 사용한다면, p만 변경될 것이다.
```python
import copy
q = copy.deepcopy(p)
```

````python
x = [1, 2, 3]
y = x[:]
x[0] = 99
print(x, y)

p = [1, 2, [3, 4]]
q = p[:]
p[2][0] = 999
print(p, q)
````

### 문제 9

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..

```python
d = {"a": 1, "b": 2, "c": 3}
k = d.keys()
d["d"] = 4
print(list(k))

# 아래 줄이 실행된다면 어떤 일이 일어나는지(출력/에러 포함) 정확히 쓰세요.
# for key in d:
#     d[key + "_x"] = d[key]
```

dict_keys class는 특별하게 사전에 대한 live view를 제공한다. 그렇기 때문에 d에서 변경점은 k에서도 live로 볼 수 있는 것이다.
만약 live view 대신 snapshot을 원한다면,
```python
k=list(d.keys())
```
위와 같이하면 k에 영향을 주지 않는 snap shot이 가능하다.

for문을 시행하면, key의 값에 _x를 더해서 새로운 key를 만들어낸다. 이는 이전의 값을 변화시키는 것은 아니기에 가능하다. 하지만, python이 iterable을 읽고 있는 와중에 사전의 사이즈를 바꾸게 된다면, 사고를 방지하기 위해서 오류를 반환한다.

````python
d = {"a": 1, "b": 2, "c": 3}
k = d.keys()
d["d"] = 4
print(list(k))

for key in d:
    d[key + "_x"] = d[key]
````

### 문제 10

다음 코드의 출력 결과를 정확히 쓰세요. 그리고 그 이유를 설명하세요..

```python
d1 = {"x": [1, 2], "y": [3, 4]}
d2 = d1.copy()
d1["x"].append(999)
print(d1)
print(d2)

d3 = {"x": (1, 2), "y": (3, 4)}
d4 = d3.copy()
# 아래 줄이 실행된다면 어떤 일이 일어나는지(출력/에러 포함) 정확히 쓰세요.
# d3["x"] += (999,)
print(d3)
print(d4)
```

위의 예시에서는 shallow copy의 한 종류로 d2는 복사된 데이터를 가리키고 있지만, 안에 리스트의 내용은 공유를 하고 있다. 그렇기에 d1 리스트 내용이 바뀌면서 d2의 리스트도 바뀐내용을 공유하는 것이다.
보통 +=는 __iadd__로서 in-place modification을 시행하지만, 튜플은 immutable이기 때문에 __iadd__가 __add__로서 역할을 해서 새로운 튜플을 만들어내게 된다. 결국 d3와 d4의 값은 다르게 되는 것이다.

````python
d1 = {"x": [1, 2], "y": [3, 4]}
d2 = d1.copy()
d1["x"].append(999)
print(d1)
print(d2)

d3 = {"x": (1, 2), "y": (3, 4)}
d4 = d3.copy()
d3["x"] += (999,)
print(d3)
print(d4)
````

````python
price_table = {
    "A100": 12000,
    "B200": 3500,
    "C300": 19900,
    "D400": 55000,
    "E500": 9900,
}

coupon_rules = {
    "A100": 10,   # 10% off
    "C300": 15,
    "E500": 5,
}

shipping_policy = {
    "fee": 3000,
    "free_over": 50000,
}

items = [("A100", 2), ("B200", 5), ("C300", 1), ("E500", 3)]
invoice = {"lines": []}

subtotal_after_discount = 0

for code, qty in items:
    # TODO 1: unit_price 가져오기
    unit_price = price_table.get(code, 0)

    # TODO 2: line_subtotal 계산 (unit_price * qty)
    line_subtotal = unit_price * qty

    # TODO 3: discount_rate 가져오기 (없으면 0)
    discount_rate = coupon_rules.get(code, 0)

    # TODO 4: line_total 계산 (정률할인 적용 후 정수 원 단위로 반올림/버림 없이 정확히 계산)
    # 힌트: 퍼센트 계산은 소수 처리 주의
    line_total = (100-discount_rate) * line_subtotal / 100

    # TODO 5: invoice["lines"]에 튜플 추가
    invoice["lines"].append((code, qty, unit_price, line_subtotal, discount_rate, line_total))

    # TODO 6: subtotal_after_discount 누적
    subtotal_after_discount += line_total


# TODO 7: 배송비 계산
if subtotal_after_discount >= shipping_policy["free_over"]:
    shipping_fee = 0
else:
    shipping_fee = shipping_policy["fee"]

# TODO 8: grand_total 계산 및 invoice에 저장
invoice["subtotal_after_discount"] = subtotal_after_discount
invoice["shipping_fee"] = shipping_fee
invoice["grand_total"] = subtotal_after_discount + shipping_fee

print(invoice)
````

````python
sales_logs = [
    ("S1", "coffee",  4500),
    ("S1", "dessert", 7000),
    ("S1", "coffee",  3000),
    ("S2", "coffee",  8000),
    ("S2", "meal",    12000),
    ("S2", "dessert", 12000),
    ("S3", "meal",    5000),
    ("S3", "coffee",  5000),
    ("S3", "meal",    5000),
    ("S1", "meal",    9000),
    ("S2", "coffee",  2000),
    ("S3", "dessert", 1000),
    ("S3", "dessert", 4000),
    ("S1", "dessert", 2000),
    ("S2", "meal",    1000),
    ("S2", "dessert", 1000),
    ("S1", "coffee",  2500),
    ("S1", "meal",    1000),
    ("S3", "coffee",  2000),
    ("S2", "coffee",  3000),
    ("S3", "meal",    2000),
]

sales_by_store = {}
top_category_by_store = {}

for store, category, amount in sales_logs:
    if store not in sales_by_store:
        sales_by_store[store] = {}

    if category not in sales_by_store[store]:
        sales_by_store[store][category]=0
    sales_by_store[store][category] += amount

top_category_by_store = {}

for store, category_totals in sales_by_store.items():
    max_category = None
    max_amount = 0

    for category, amount in category_totals.items():
        if max_category is None or amount > max_amount:
            max_category = category
            max_amount = amount
    top_category_by_store[store] = (max_category, max_amount)


print(sales_by_store)
print(top_category_by_store)
````
