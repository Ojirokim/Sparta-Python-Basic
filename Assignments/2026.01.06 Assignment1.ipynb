input()의 결과는 항상 string이기 때문이다.

```python
# 3-4번, 아래 코드를 완성하세요.
data = input()

# 합 / 차 / 곱 출력
a,b =data.split()
a, b = int(a), int(b)
print(f"합: {a+b}\n차: {a-b}\n곱: {a*b}")
```

```python
# 3-5번, 아래 코드를 완성하세요.
data = input()

# try / except 사용
try:
    a,b =data.split()
    a, b = int(a), int(b)
    print(f"합: {a+b}\n차: {a-b}\n곱: {a*b}")
except ValueError:
    print("형식이 잘못되었습니다")
```

```python
# 3-6번, 아래 코드를 완성하세요.
data = input()
a,b =data.split(',')
a,b = int(a),int(b)
print(a*b)
```

```python
# 3-7번, 아래 코드를 완성하세요.
num = input()
num=int(num.replace(',',''))
print(num*2)
```

int() 함수는 float를 받아서 int로 변경하던지 string을 받아서 int로 변경해주는 것인데, float이면서 string인 경우는 오류가 난다.

```python
# 4-1번
data = input()
a,b,c = map(int,data.split())
print(a+b+c)
```

```python
# 4-2번
data = input()
a= data.split(",")
nums = [int(x) for x in a]
print(f"{sum(nums)/len(nums):.2f}")
```

```python
# 4-3번
data = input()
a= data.split('#')
nums= [int(x) for x in a]
print(max(nums))
```
