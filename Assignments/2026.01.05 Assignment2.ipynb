#### ✅ 좋은 변수명 예제(3개)

```python
# 예제 1: 나쁜 변수명
a = 3
b = 5
c = a * b
print("결과:", c)
# 예제 2: 좋은 변수명
width = 3
height = 5
area = width * height
print("넓이:", area)
# 예제 3: 의미가 있으면 디버깅(고치기)도 쉬워집니다
student_name = "적토마"
math_score = 90
english_score = 80
total_score = math_score + english_score
print(student_name, "총점:", total_score)
```

#### ✅ `int()` 예제 (4개)

```python
print(int("10") + 5)
print(int("0") * 100)
print(int("-7") - 3)
print(int("   42   "))  # 앞뒤 공백은 보통 허용
```

#### ✅ `float()` 예제 (4개)

```python
print(float("3.5") + 1.2)
print(float("0.1") * 10)
print(float("-2.75") / 2)
print(float("1e3"))  # 1000.0 (과학표기)
```

#### ✅ `str()` 예제 (4개)

```python
n = 50
print("점수는 " + str(n) + "점입니다.")
# f-string을 쓰면 str()를 직접 안 써도 되는 경우가 많습니다(참고).
n = 50
print(f"점수는 {n}점입니다.")
# 문자열 반복 만들기
stars = "*" * 10
print(stars)
# 숫자 리스트를 문자열로 바꿔 join하기(맛보기)
nums = [1, 2, 3]
text = ",".join(map(str, nums))
print(text)
```

#### 산술 ✅ 예제 (6개)

```python
a = 12
b = 5
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
# 괄호 먼저 계산
print((2 + 3) * 4)
# 음수도 가능
temp = -3
print(temp * 10)
# 실수 계산
x = 1.5
y = 2.0
print(x * y)
print("Py" + "thon")
```

#### 다중할당 ✅ 예제 (6개)

```python
a, b = 1, 2
print(a, b)
x, y, z = 10, 20, 30
print(x, y, z)
# swap(바꾸기)
a, b = 100, 200
a, b = b, a
print(a, b)
# 리스트 같은 것도 나눠 담을 수 있습니다(맛보기)
p, q = [7, 9]
print(p, q)
# input과 같이 쓰는 패턴(직접 입력: 예 3 7)
a, b = map(int, input("정수 2개(공백): ").split())
print("합:", a + b)
# (헷갈림) 변수 개수와 값 개수가 다르면 에러
# a, b = 1, 2, 3  # ValueError
print("개수 불일치 예시는 주석으로 막아두었습니다.")
```

#### 동일값 할당✅ 예제 (4개)

```python
x = y = z = 0
print(x, y, z)
math_score = english_score = science_score = 0
print(math_score, english_score, science_score)
a = b = 10
a = a + 5
print("a:", a, "b:", b)
a = b = "hi"
print(a, b)
```

### 미니 프로젝트: “용돈 계산기” (변수 총정리)

```python
allowance = 50000
snack = 12000
transport = 15000
game = 8000

total_spent = (snack+transport+game)
print(total_spent)

remaining = allowance-total_spent
print(remaining)
```
