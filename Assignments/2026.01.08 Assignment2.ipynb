```python
### 필수 1번
def printing():
    for x in range(1,21):
        if x%3==0 and x%6!=0:
            print(x)
        else:
            continue
printing()
```

```python
### 필수 2번
def printing():
    a=[]
    for x in range(1,11):
        if x%2!=0:
            a.append(x)
            print(x)
        else:
            continue
    print(f'출력된 숫자 개수: {len(a)}')
printing()
```

```python
### 필수 3번
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
### 필수 4번
def printing():
    a=input()
    b= len(a)
    for x in range(0,b):
        if x%2==0:
            print(a[x])
printing()
```

```python
### 필수 5번
def printing():
    for x in range(1,3):
        for y in range(1,4):
            print(x,y)
printing()
```

```python
### 필수 6번
def input_add():
    try:
        x=1
        num_list=[]
        while x>0:
            num= int(input())
            if num!=0:
                num_list.append(num)
                print(f"숫자 입력: {sum(num_list)}")
            elif num==0:
                print(f"숫자 입력: 0")
                break
            x+=1
        print(f"총합: {sum(num_list)}")
    except ValueError:
        print("숫자 값을 다시 입력하세요")

input_add()
```

```python
### 필수 7번
def print_until():
    n = int(input())
    print(f"입력: {n}")
    print("출력")
    for i in range(1, n+1):
        print(i)
print_until()
```

```python
### 필수 8번
def reverse_print():
    x= int(input())
    print(f"입력: {x}")
    print("출력")
    for i in range(x,0,-1):
        print(i)
reverse_print()
```

```python
### 필수 9번
def print_until():
    n = int(input())
    a=0
    print(f"입력: {n}")
    for i in range(1, n+1):
        a=a+i
    print(f"출력: 1부터 {n}까지의 합은 {a}")
print_until()
```
