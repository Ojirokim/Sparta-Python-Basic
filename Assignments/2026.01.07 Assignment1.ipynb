```python
## 실습문제2
def calculate_bmi():
    try:
        kg = float(input("Enter weight in kg: "))
        cm = float(input("Enter height in cm: "))

        if kg <= 0 or cm <= 0:
            print("Error: Weight and height must be positive numbers.")
            return

        meters = cm / 100
        bmi = kg / (meters ** 2)
        
        if bmi < 18.5:
            print(f"BMI: {bmi:.1f} → 저체중")
        elif 18.5 <= bmi < 23:
            print(f"BMI: {bmi:.1f} → 정상")
        elif 23 <= bmi < 25:
            print(f"BMI: {bmi:.1f} → 과체중")
        else:
            print(f"BMI: {bmi:.1f} → 비만")

    except ValueError:
        print("Error: Please enter valid numeric values.")

calculate_bmi()
```

```python
## 실습문제3
def triangle():
    try:
        a= float(input())
        b= float(input())
        c= float(input())
        if a<=0 or b<=0 or c<=0:
            print("양수를 적어주세요")
            return
        if a+b>c and a+c>b and b+c>a:
            if a**2+b**2==c**2:
                print("직각삼각형")
            elif a**2+c**2==b**2:
                print("직각삼각형")
            elif c**2+b**2==a**2:
                print("직각삼각형")
            elif a==b and b==c:
                print("정삼각형")
            elif a==b and b!=c:
                print("이등변삼각형")
            elif a==c and b!=c:
                print("이등변삼각형")
            elif b==c and a!=c:
                print("이등변삼각형")
            elif a!=b and b!=c and a!=c:
                print("일반 삼각형")
        else:
            print("삼각형이 아닙니다")
    except ValueError:
        print("제대로된 값을 적으시오")
triangle()
```

```python
## 실습문제3 AI 코드수정
def triangle():
    try:
        sides = [float(input()) for _ in range(3)]
        if any(s <= 0 for s in sides):
            print("양수를 적어주세요")
            return

        sides.sort()  # a <= b <= c
        a, b, c = sides

        if a + b <= c:
            print("삼각형이 아닙니다")
            return

        # float 비교는 오차가 있을 수 있어서 tolerance 사용
        tol = 1e-9
        if abs(a*a + b*b - c*c) < tol:
            print("직각삼각형")
        elif a == b == c:
            print("정삼각형")
        elif a == b or b == c:  # (정렬했으니 a==c인 경우는 정삼각형밖에 없음)
            print("이등변삼각형")
        else:
            print("일반 삼각형")

    except ValueError:
        print("제대로된 값을 적으시오")

triangle()
```
