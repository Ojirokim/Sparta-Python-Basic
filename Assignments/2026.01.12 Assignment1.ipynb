### 실습 문제 1 : 리스트의 합 `list_sum`

**문제 설명**

- 정수로만 이루어진 리스트 `nums`를 매개변수로 받아 **모든 요소의 합**을 반환하는 함수를 만드세요.
- 리스트가 비어 있다면 `0`을 반환해야 합니다.

```python
def list_sum(nums):
    return sum(nums)
print(list_sum([1, 2, 3, 4]))
print(list_sum([]))
```

### 실습 문제 2 : 단어 빈도 세기 함수 `word_count`

**문제 설명**

- 문자열을 매개변수로 받아, 공백을 기준으로 단어를 분리합니다.
- 각 단어가 몇 번 나왔는지 **딕셔너리** 형태로 반환하세요.
- 대소문자는 구분하지 않으며, 모두 소문자로 변환한 뒤 계산하세요.

```python
def word_count(word):
    word_divide=word.lower().split()
    word_count={}
    distinct_word= sorted(set(word_divide))
    for i in distinct_word:
        word_count[i]=word_divide.count(i)
    return word_count
print(word_count("Apple banana apple Orange orange banana apple"))
```

### 실습 문제 3 : 학점 계산 함수 `grade_report`

**문제 설명**

- 학생 이름과 점수가 들어있는 딕셔너리를 매개변수로 받으세요.
- 점수를 기준으로 **학점(A, B, C, D, F)**을 매겨서 새 딕셔너리로 반환하세요.
- 학점 기준은 다음과 같습니다.
| 점수 범위 | 학점 |
| --- | --- |
| 90 이상 | A |
| 80~89 | B |
| 70~79 | C |
| 60~69 | D |
| 60 미만 | F |

```python
def grade_report(scores):
    grade_info={}
    for name, score in scores.items():
        if score>=90:
            grade_info[name]="A"
        elif score>=80:
            grade_info[name]="B"
        elif score>=70:
            grade_info[name]="C"
        elif score>=60:
            grade_info[name]="D"
        else:
            grade_info[name]="F"
    return grade_info
scores = {"Alice": 85, "Bob": 92, "Charlie": 67, "Dave": 58}
print(grade_report(scores))
```
