### 문제 1

두 개의 숫자를 받아서 **더 큰 값을 반환하는 함수** `bigger(a, b)`를 만들어 보세요.

예를 들어 `bigger(3, 5)`를 실행하면 `5`가 반환되어야 합니다.

````python
def bigger(a,b):
    return max(a,b)
print(bigger(3,5))
````

### 문제 2

### 문제 5

아래 함수는 리스트의 첫 번째 값을 0으로 바꾸려고 합니다.

하지만 실행 결과를 보면 원래 리스트가 바뀌지 않습니다.

그 이유를 설명하고, 리스트가 실제로 바뀌도록 코드를 수정해 보세요.

```python
def change_first(x):
    x = [0, 2, 3]

nums = [1, 2, 3]
change_first(nums)
print(nums)

```

````python
def change_first(x):
    x[0] = 0
    return x
nums = [1, 2, 3]
change_first(nums)
# x는 새로운 list를 가리키며, nums의 값은 실제로 바뀌지 않는다.
````

### 문제 6

하지만 출력 결과가 예상과 다릅니다.

왜 이런 결과가 나오는지 설명해 보세요.

```python
def add_one(n):
    n = n + 1

x = 10

print(x)
```


### 문제 8

아래 함수는 문자열을 모두 대문자로 바꾸려고 만든 함수입니다.

하지만 `print` 결과가 기대와 다릅니다.

**왜 문자열이 바뀌지 않는지** 설명하세요.

```python
def make_upper(s):
    s.upper()

text = "hello"
make_upper(text)
print(text)
```

````python
def make_upper(s):
    s.upper()

text = "hello"
make_upper(text)
print(text)
# string은 immutable이다. 그렇기 때문에 str.upper()를 하게 되면 새로운 string을 만드는데, 그 결과를 return하지도 않았고, 특정 정의변수에 넣지도 않았기 때문이다.
````

### 문제 9

`math` 모듈을 사용하여 **원의 넓이**를 계산하는 프로그램을 만들어 보세요.

사용 방법:

- 반지름을 입력받는다.
- `math.pi`를 사용하여 원의 넓이(π × r × r)를 계산하는 함수를 만든다.
- 계산된 넓이를 출력한다.

예시 상황:

### 문제 10

`random` 모듈을 사용하여 **주사위 프로그램**을 만들어 보세요.

요구사항:

- 1부터 6 사이의 숫자를 하나 랜덤으로 뽑는 함수를 만든다.
- 뽑힌 숫자를 화면에 출력한다.
- 이 함수를 3번 호출해서 주사위를 3번 굴린 결과를 출력한다.

출력 예시:

```
주사위 결과: 4
주사위 결과: 1
주사위 결과: 6
```

````python
import random
def randoo():
    random_num= random.randint(1,6)
    return print(f"주사위 결과: {random_num}")
randoo()
randoo()
randoo()
````
