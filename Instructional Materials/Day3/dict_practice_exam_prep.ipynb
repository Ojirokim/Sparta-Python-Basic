## ðŸŸ¢ Level 1 â€” Problem 1: Login Attempts Summary

## ðŸŸ¡ Level 2 â€” Problem 3: Top Product Per Store (Tie-breaking)

Each log is `(store, product, quantity)`.

### Requirements
- Build `product_by_store` as `{store: {product: total_quantity}}`
- Build `top_product_by_store` as `{store: (top_product, total_quantity)}`
- If quantities tie, choose **lexicographically smaller product**


```python
logs = [
    ("S1", "apple", 10),
    ("S1", "banana", 10),
    ("S1", "apple", 5),
    ("S2", "orange", 8),
    ("S2", "apple", 8),
]

product_by_store = {}
top_product_by_store = {}
def product_into_dic():
    for store, product, quantity in logs:
        if store not in product_by_store.keys():
            product_by_store[store]={}
            if product not in product_by_store[store].keys():
                product_by_store[store][product]=quantity
            else:
                product_by_store[store][product]+=quantity
        else:
            if product not in product_by_store[store].keys():
                    product_by_store[store][product]=quantity
            else:
                product_by_store[store][product]+=quantity
product_into_dic()
print(product_by_store)
def top_product_dic():


    for store, product_quantity in product_by_store.items():
        maxim = -1
        product_name =None
        for product, quantity in product_quantity.items():
            if maxim<quantity:
                maxim=quantity
                product_name=product
            elif maxim==quantity:
                if product_name>product:
                    maxim=quantity
                    product_name=product
        top_product_by_store[store]=(product_name,maxim)

top_product_dic()
print(top_product_by_store)
```

## ðŸ”µ Level 3 â€” Problem 5: Order Fulfillment (All-or-nothing)

Each order has multiple `(item, quantity)` lines.

### Requirements
- An order is accepted **only if all items can be fulfilled**
- Inventory must only change if order is accepted
- Build `accepted_orders` and `rejected_orders`


```python
inventory = {
    "apple": 10,
    "banana": 5,
    "orange": 8,
}

orders = [
    ("o1", [("apple", 3), ("banana", 2)]),
    ("o2", [("banana", 4), ("orange", 5)]),
    ("o3", [("apple", 5), ("orange", 10)]),
]

orders_dict={}
accepted_orders = []
rejected_orders = []

# mental road 1: create order dict
for order, specifics in orders:
    orders_dict[order]={}
    for product, quantity in specifics:
        orders_dict[order][product]=quantity

# mental road 2: compare that dict to inventory to get accepted_orders and rejected_orders
for order, specifics in orders_dict.items():
    okay_check=True
    for product, quantity in specifics.items():
        if inventory[product]>=orders_dict[order][product]:
            okay_check=True
        else:
            okay_check=False
    if okay_check:
        accepted_orders.append(order)
        for product, quantity in specifics.items():
            inventory[product]-=quantity
    else:
        rejected_orders.append(order)


print(orders_dict)
print(accepted_orders)
print(rejected_orders)
print(inventory)
```
