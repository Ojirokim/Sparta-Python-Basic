## ðŸŸ¢ Level 1 â€” Problem 1: Login Attempts Summary

You are given login logs. Each log is `(user_id, status)` where status is `"OK"` or `"FAIL"`.

### Requirements
- Build `attempts_by_user` as `{user_id: [total, success, fail]}`
- Build `locked_users` as a list of users whose `fail >= lock_threshold`
- `locked_users` should have **no duplicates**, sorted ascending


```python
logs = [
    ("u01", "OK"),
    ("u01", "FAIL"),
    ("u02", "FAIL"),
    ("u01", "FAIL"),
    ("u02", "OK"),
    ("u02", "FAIL"),
    ("u02", "FAIL"),
    ("u03", "OK"),
]

lock_threshold = 3

attempts_by_user = {}
locked_users = []

# TODO: implement
for key, value in logs:
    if key not in attempts_by_user:
        attempts_by_user[key]= [0,0,0]
    if value=="FAIL":
        attempts_by_user[key][0] +=1
        attempts_by_user[key][2] +=1
    else:
        attempts_by_user[key][0] +=1
        attempts_by_user[key][1] +=1
for key, value in attempts_by_user.items():
    if value[2]>= lock_threshold:
        locked_users.append(key)
locked_users.sort()

print(attempts_by_user)
print(locked_users)
```

## ðŸŸ¡ Level 2 â€” Problem 3: Top Product Per Store (Tie-breaking)

Each log is `(store, product, quantity)`.

### Requirements
- Build `product_by_store` as `{store: {product: total_quantity}}`
- Build `top_product_by_store` as `{store: (top_product, total_quantity)}`
- If quantities tie, choose **lexicographically smaller product**


```python
logs = [
    ("S1", "apple", 10),
    ("S1", "banana", 10),
    ("S1", "apple", 5),
    ("S2", "orange", 8),
    ("S2", "apple", 8),
]

product_by_store = {}
top_product_by_store = {}

# TODO 1: aggregate totals into product_by_store
for store, product, quantity in logs:
    if store not in product_by_store:
        product_by_store[store]={}
    if product not in product_by_store[store]:
        product_by_store[store][product]=0
    product_by_store[store][product] +=quantity
for store, product_dic in product_by_store.items():
    best_product = None
    best_qty = -1

    for product, total_quantity in product_dic.items():
        if (best_product is None or
            total_quantity > best_qty or
            (total_quantity == best_qty and product < best_product)):
            best_product = product
            best_qty = total_quantity

    top_product_by_store[store] = (best_product, best_qty)
print(product_by_store)
print(top_product_by_store)
```
